```python
import matplotlib.pyplot as plt
import numpy as np
import fdicdata as fdic

# Fetch the data
data = fdic.getSummary(states=["Florida", "Rhode Island", "Texas"], date_range=["2020", "2024"], fields=["DEP", "ASSET"])

# Extract states, DEP, and ASSET values
years = data['YEAR'].unique()
states = data['STNAME'].unique()

# Group data by year and state, then sum DEP and ASSET
grouped_data = data.groupby(['STNAME', 'YEAR'])[['DEP', 'ASSET']].sum().reset_index()

# Create the plot
fig, axs = plt.subplots(10, 6, figsize=(12, 8))  # Adjust figsize as needed
axs = axs.flatten()  # Flatten the array of Axes objects

# Set up the plot
x = np.arange(len(states))  # the label locations
width = 0.35  # the width of the bars

ax_index = 0  # Keep track of current Axes object index

for i, year in enumerate(years):
    year_data = grouped_data[grouped_data['YEAR'] == year]
    dep_values = year_data['DEP'].values
    asset_values = year_data['ASSET'].values

    bar_offset = i * width

    # Access the correct Axes object using ax_index
    ax = axs[ax_index]

    rects1 = ax.bar(x - width/2 + bar_offset, dep_values, width, label=f'DEP ({year})')
    rects2 = ax.bar(x + width/2 + bar_offset, asset_values, width, label=f'ASSET ({year})')

    ax_index += 1  # Move to the next Axes object

# Add some text for labels, title and custom x-axis tick labels, etc.
# Only set labels for the first Axes object
axs[0].set_ylabel('Values')
axs[0].set_title('DEP and ASSET by State and Year')
axs[0].set_xticks(x)
axs[0].set_xticklabels(states)
axs[0].legend(loc='upper left', bbox_to_anchor=(1, 1))

# Hide unused subplots
for i in range(ax_index, len(axs)):
    axs[i].axis('off')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import requests

#original code from https://github.com/Visbanking/fdicdataPy/blob/main/fdicdata/getFailures.py
#Modified to showcase more of the data it's original code was gathering.

def getFailures(fields, date_range=None, name=None, limit=10000):
    import io
    assert fields is not None, "Fields cannot be empty"

    if date_range is not None:
        assert len(date_range) == 2, "Date range must have two values"
        assert len(date_range[0]) == len(date_range[1]) == 4, "Date range values must be of length 4"
        assert date_range[0] <= date_range[1], "Invalid date range values"

    url = "https://banks.data.fdic.gov/api/failures?"

    filters = []

    if date_range is not None:
        filters.append(f"FAILYR%3A%5B%22{date_range[0]}%22%20TO%20%22{date_range[1]}%22%5D")

    if name is not None:
        filters.append(f"NAME%3A%22{name}%22")

    if filters:
        url += "filters=" + "%20AND%20".join(filters) + "&"
    else:
        url += "filters=FAILYR%3A%5B%221900%22%20TO%20%222300%22%5D&"

    url += f"fields=CERT%2CNAME%2CFAILDATE%2C{'%2C'.join(fields)}&limit={limit}&format=csv&download=false&filename=data_file"

    try:
        response = requests.get(url)
        df = pd.read_csv(io.StringIO(response.text))
        df = df.assign(FAILDATE=pd.to_datetime(df["FAILDATE"], format="%m/%d/%Y")).drop("ID", axis=1)
        return df

    except Exception as e:
        print(f"ERROR: {str(e)}")
        return None

print(getFailures('CERT'))
```

```python
import io
import pandas as pd
import requests

# example: print(getHistory(CERT_or_NAME=3850, fields=["INSTNAME", "CERT", "PCITY", "PSTALP", "PZIP5"], CERT=True, limit=10))

def getHistory(CERT_or_NAME, fields, CERT=True, limit=10000):
    assert CERT_or_NAME is not None and fields is not None, "CERT_or_NAME and fields cannot be empty"

    url = f"https://banks.data.fdic.gov/api/history?"
    if CERT:
        url += f"filters=CERT%3A%20{CERT_or_NAME}"
    else:
        url += f"search=NAME%3A%20%27{CERT_or_NAME.replace(' ', '%20')}%27"
    url += f"&fields=CERT%2C{('%2C'.join(fields))}&limit={limit}&format=csv&download=false&filename=data_file"

    try:
        response = requests.get(url)
        df = pd.read_csv(io.StringIO(response.text))
        df = df.drop(["ID"], axis=1)
        return df
    except Exception as e:
        print(f"ERROR: {str(e)}")
        return None

print(getHistory(CERT_or_NAME=3850, fields=["INSTNAME", "CERT", "PCITY", "PSTALP", "PZIP5"], CERT=True, limit=10))
```
